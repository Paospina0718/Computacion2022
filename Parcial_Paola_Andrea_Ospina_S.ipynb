```python
import pandas as pd
import pingouin as pg
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
import seaborn as sns
```

```python
df1=pd.read_excel("https://docs.google.com/spreadsheets/d/e/2PACX-1vRI-8pRn-q_JirOhj8vtgdXJe7oROV2ERLhc2-_-89263IUgb8BuS1mbeP_HXf2qg/pub?output=xlsx")
df1.head
```

```python
#Analisis de varianza
#Metodo 1
pg.anova(data=df1,dv="CBD",between="Accesion")
```

```python
#Metodo 2
formula=ols("CBD~(Accesion)",data=df1).fit()
anova_lm(formula)
```

Supuestos

```python
from pingouin.distribution import homoscedasticity
from pingouin.distribution import normality
#Normalidad
pg.normality(formula.resid,method="shapiro")
```

```python
#Homocedasticidad
df1["resid"]=formula.resid
pg.homoscedasticity(dv="resid", group="Accesion",data=df1)
```
